```python
import pandas as pd
import numpy as np
import joblib
import pickle
```

```python
def prediction_churn(input_data):
    import numpy as np, joblib, pickle

    model = joblib.load("xgboost_churn_model.pkl")
    scaler = pickle.load(open("scalar.pkl", "rb"))
    encoders = pickle.load(open("encoders.pkl", "rb"))

    # encode Department ONLY
    input_data["Department"] = encoders["Department"].transform(
        [input_data["Department"]]
    )[0]

    feature_order = [
        "Age", "Tenure", "Salary", "Overtime Hours",
         "Satisfaction Level",
        "Promotions",
        "Manager Feedback Score", "Department"
    ]

    X = np.array([input_data[f] for f in feature_order]).reshape(1, -1)
    X = scaler.transform(X)

    pred = model.predict(X)[0]
    prob = model.predict_proba(X)[0][1]


    if prob > 0.5:
        print("The Employee is likely to churn")
    else:
        print("The Employee is not likely to churn")
```

```python
sample_input = {
    "Age": 35,
    "Tenure": 5,
    "Salary": 60000,
    "Overtime Hours": 12,
    "Satisfaction Level": 0.4,
    "Promotions": 1,
    "Manager Feedback Score": 3,
    "Department": "IT"
}

print(prediction_churn(sample_input))

```
